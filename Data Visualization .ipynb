## Data Visualization using Seaborn

Seaborn is a library that uses Matplotlib underneath to plot graphs

```python
#importing the required libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.lineplot(x="SepalLengthCm",y="SepalWidthCm" , data=df, errorbar=None)
plt.show()
```

```python
sns.lineplot(x="SepalLengthCm", y="SepalWidthCm" , data=df, hue="Species",errorbar=None)
plt.show()
```

```python
#continous and categorical feature
sns.lineplot(x="SepalWidthCm", y="Species" , data=df, hue="Species",errorbar=None)
plt.show()
```

```python
sns.lineplot(x="PetalLengthCm", y="Species" , data=df, hue="Species",errorbar=None)
plt.show()
```

```python
#Scatter plot
sns.scatterplot(x="SepalLengthCm", y="SepalWidthCm", data=df, hue="Species")
plt.title("Scatter plot")
plt.xlabel("Y")   #Custom X axis label
plt.show()
```

```python
#Bar plot
sns.barplot(x="Species" , y="SepalLengthCm" , data=df, estimator="mean")
plt.show()
```

```python
#Count Plot
sns.countplot(x="Species", data=df)
plt.show()
```

```python
#Box plot or whisker plot
sns.boxplot(x="SepalWidthCm", data=df)
plt.show()
```

```python
sns.boxplot(x="Species" , y="SepalWidthCm" , data=df)
plt.show()
#Checking outliers in each category with respect to sepal width
```

```python
#Violin plot
#It provides a more advanced visualization than box plot
sns.violinplot(x="Species" , y="SepalWidthCm" , data=df)
plt.show()
```

```python
#Histogram
sns.histplot(x="PetalWidthCm", data=df, bins=4)
plt.show()
```

```python
sns.histplot(x="PetalLengthCm", data=df, hue="Species", kde=True) #kernel density estimation
plt.show()
#kde is approximation of histogram to density curve
```

## Correlation

```python
#Comparing two values
```

```python
df.columns
```

```python
tc=df.drop(columns = ["Id", "Species"], axis=1)
```

```python
tc = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
```

```python
##Heat map
## Heatmap is defined as a graphical representation of data using colors to visualize the value of the matrix.
## In this, to represent more common values or higher activities brighter colors basically reddish colors are used and 
## to represent less common or activity values, darker colors are preferred

# correlation between the different parameters 
tc = tc.corr()
  
tc
```

```python
sns.heatmap(tc, annot = True)
plt.show()
```
